**Question 04**

Plot the histogram of each of these features in a 1x4 layout.

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 4, figsize=(20, 5))
#fig is the entire figure object.
#axs is a list of 4 individual axes (where each histogram will be drawn).
#figsize=(20, 5) sets the size of the entire figure to 20 inches wide and 5 inches tall.

for i, col in enumerate(new_df.columns): #Loops through each column in new_df.enumerate gives both:i: the index (0, 1, 2, 3)col: the column name (e.g., "some_column")
    axs[i].hist(new_df[col], bins=30, color='skyblue', edgecolor='black')
    axs[i].set_title(col)
    axs[i].set_xlabel(col)
    axs[i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
