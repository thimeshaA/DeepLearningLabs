**Question 02**

 Create a random 100000x1 array from the same exponential distribution and visualize the distribution histogram along with uniform and normal distributions. Use the code given below and do the necessary changes. Change the number of bins and the plot view to make the visualization better.


---



| Feature                    | **Exponential**                                                                              | **Uniform**                                                                              | **Normal (Gaussian)**                                                                    |
| -------------------------- | -------------------------------------------------------------------------------------------- | ---------------------------------------------------------------------------------------- | ---------------------------------------------------------------------------------------- |
| **Shape**                  | Right-skewed (long tail on the right)                                                        | Flat line                                                                                | Bell curve (symmetric)                                                                   |
| **Range of values**        | \[0, ∞)                                                                                      | \[0, 1) by default                                                                       | $-∞, ∞$                                                                                  |
| **Mean (μ)**               | Equal to the scale parameter                                                                 | Middle of the interval (e.g., 0.5)                                                       | Defined by parameter (default 0)                                                         |
| **Standard deviation (σ)** | Equal to the mean (scale)                                                                    | Depends on range                                                                         | Defined by parameter (default 1)                                                         |
| **Where it is used**       | Time between random events (e.g., arrivals)                                                  | Random selection with equal probability                                                  | Natural phenomena, heights, test scores                                                  |
| **Skewness**               | Positive (right skew)                                                                        | 0 (perfectly symmetric)                                                                  | 0 (perfectly symmetric)                                                                  |
| **Kurtosis**               | 6                                                                                            | -1.2                                                                                     | 3 (mesokurtic — normal)                                                                  |
| **Typical Function**       | `np.random.exponential(scale=1.0)`                                                           | `np.random.rand()`                                                                       | `np.random.randn()`                                                                      |
| **PDF Formula**            | $f(x) = \lambda e^{-\lambda x}$                                                              | $f(x) = \frac{1}{b - a}$                                                                 | $f(x) = \frac{1}{\sqrt{2\pi}} e^{-x^2/2}$                                                |



```python

import matplotlib.pyplot as plt

data_exp = np.random.exponential(size=100000)
data_uniform = np.random.rand(100000)
data_normal = np.random.randn(100000)

plt.hist(data_exp, density=True, bins=100, histtype="step", label="Exponential", color="red")
plt.hist(data_uniform, density=True, bins=100, histtype="step", label="Uniform", color="blue")
plt.hist(data_normal, density=True, bins=100, histtype="step", label="Normal", color="green")
# density=True: Scales the area under the histogram to 1 (makes it a probability density).
# bins=100: Splits the data into 100 intervals for more detail.
# histtype="step": Makes a line-style histogram (clearer than bars when multiple plots overlap).(bar, barstacked, step, stepfilled)
# label and color: Help differentiate each distribution.

plt.legend(loc="upper right")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.axis([-2.5, 2.5, 0, 1.1])
plt.show()

```

**Question 03**

Using matplotlib, plot the Z = X**2 + Y**2 in a 3D plot. Limit the x and y values to [-5, 5].


---



```python
from mpl_toolkits.mplot3d import Axes3D

# Step 1: Create x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Step 2: Create meshgrid
X, Y = np.meshgrid(x, y)

# Step 3: Calculate Z
Z = X**2 + Y**2

# Step 4: Plot
fig = plt.figure(1, figsize = (6, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
# cmap='viridis' applies a nice color map that shows depth (low values are dark blue, high are yellow-green).


ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z = X² + Y²')
ax.set_title("3D Surface Plot: Z = X² + Y²")

plt.show()
```

**Question 04**

Calculate the pearson standard coefficient and spearman rank correlation for the [HP,Attack, Defense, Sp. Ark, Sp. Def, Speed] features in the seaborn tutorial.ipynb. Visualize the results using heatmaps. Make sure the values are displaced in the heatmap as well.


---



Pearson correlation : measures linear relationship.

Spearman correlation : measures monotonic (rank-based) relationship.

```python
import seaborn as sns #seaborn is used to create nice-looking statistical plots like heatmaps.
import pandas as pd #pandas is used to manage the data (DataFrame)
import matplotlib.pyplot as plt

data = {
    'HP': [45, 60, 80, 70, 90],
    'Attack': [49, 62, 100, 75, 85],
    'Defense': [49, 63, 80, 70, 95],
    'Sp. Atk': [65, 80, 95, 85, 110],
    'Sp. Def': [65, 80, 85, 70, 95],
    'Speed': [45, 60, 70, 90, 100]
}

df = pd.DataFrame(data)
# df is the dataframe
features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']

corr_pearson = df[features].corr(method='pearson')
corr_spearman = df[features].corr(method='spearman')

plt.figure(figsize=(10, 4)) #(width,height)

plt.subplot(1, 2, 1)  # 1 row, 2 columns, this is the 1st plot
sns.heatmap(corr_pearson, annot=True, cmap='coolwarm')
plt.title("Pearson Correlation")

plt.subplot(1, 2, 2)  # 1 row, 2 columns, this is the 2nd plot
sns.heatmap(corr_spearman, annot=True, cmap='coolwarm')
plt.title("Spearman Correlation")

#annot=True: Writes the correlation values inside each square of the heatmap.
#cmap='coolwarm': Uses a color scale where red means strong positive correlation, blue means strong negative.

plt.tight_layout()
plt.show()
```
